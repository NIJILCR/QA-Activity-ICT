```python
x = data.drop(['address','famsize','Pstatus','reason','guardian','schoolsup', 'famsup', 'paid','nursery',
       'higher', 'internet', 'romantic','passed'],axis=1)
y = pd.DataFrame(data['passed'])
```

```python
x = pd.get_dummies(x, columns=['school','activities','sex'])

y = pd.get_dummies(y, columns=['passed'])
```

```python
x.head()
```

```python
y.head()
```

```python
x['pjob']=x['Mjob']+['Fjob']
x['freetimes']=x['freetime']+x['goout']
```

```python
x=x.drop(['Mjob','Fjob','freetime','goout'],axis=1)
```

```python
x.head()
```

```python
x.info()
```

```python
x=x.drop(['pjob'],axis=1)
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.25)
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
x_train.head()
```

# RandomForest_Model

```python
from sklearn.ensemble import RandomForestClassifier
rf1=RandomForestClassifier()
rf1.fit(x_train,y_train)
```

```python
y_pred = rf1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import f1_score,confusion_matrix
print('f1_score is:',f1_score(y_test,y_pred,average='weighted'))
print('confusion_matrix:',confusion_matrix(y_test.values.argmax(axis=1),y_pred.argmax(axis=1)))
```

# Feature importance verification

```python
pd.Series(rf1.feature_importances_,index=x.columns).sort_values(ascending = False)*100
```

```python
rf1.fit(x_train,y_train)
```

```python
rfit1= RandomForestClassifier(n_estimators=500,max_depth=10,random_state=42,criterion="entropy")
rfit1.fit(x_train,y_train)
```

```python
y_pred=rfit1.predict(x_test)
f1_score(y_test,y_pred,average='weighted')
```
